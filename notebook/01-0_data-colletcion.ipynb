## 1. Data collection
All raw datasets were manually downloaded for the period between 2021-01-01 and 2025-12-31 in order to analyze medium-term market dynamics, seasonality, and volatility patterns.  

### Data collection and dataset description
The downloaded dataset comes from two sources: 
1. The first one is GME ([Gestore del Mercato Elettrico]((https://www.mercatoelettrico.org/it-it/Home/Esiti/Elettricita/MGP/Esiti/PUN))) which is the Italian NEMO (Nominated Electricity Market Operator).  
The dataset contains the Italian PUN index (Prezzo Unico Nazionale), namely the national single electricity price calculated as the consumption-weighted average of the zonal market prices, of 1 year from January 1st to December 31st and it's composed by the following columns:
    * _Data_ - it's the date, formatted as dd/mm/yy in ascending order;
    * _Ora_ - it's the hour of the day, data type is integer that goes from 1 to 24;
    * _€/MWh_ - data type is float and it represent the price of the energy at that hour of the day by Mega Watt per hour. 
2. The second one is [TERNA Downolad Data Center](https://dati.terna.it/en/download-center). TERNA (Trasmissione Elettrica Rete Nazionale) is the Italian TSO (Transmission System Operator).   
The dataset(*) is divided in two macro-areas:  
    1. **Generation** - it represents the  quantity of energy generated and in this study it describes the supply (S) of the energy.  
    The dataset is composed by 5 files (and tables), each one for each year and the columns contained are:
        * _Date_ - date format is <code>dd/mm/yy hh:mm:ss</code> on quarter-hourly frequency;
        * _Actual Generation_ - datatype is float, and values are reported in Megawatts (MW);
        * _Primary Source_ - it represents the source of energy production and the values are Thermal_, _Wind_, _Geothermal_, _Photovoltaic_, _Self-consuption_, _Hydro_.
    2. **Load** - it represents the quantity of energy dispatched, for this study it is used as a proxy for electricity demand (D).
    The dataset is composed by 5 files (and tables), one for each year. The columns are:
        * _Date_ - date format is <code>dd/mm/yy hh:mm:ss</code> at 15-minute intervals;
        * _Total Load [MW]_ - float data type
        * _Forecast Total Load [MW]_ - for this analysis will not be taken into consideration
        * _Bidding Zone_ - it would be helpful for zonal analysis and zonal market dynamic. The zones, classified by energy transmission net prerequisites and characetristics,  are _Calabria_, _Centre-North_, _Centre-South_, _North_, _Sardinia_, _Sicily_, _South_, _Italy_.   

> (*) File naming convention:
> * For GME dataset is <code>GME-PUNi-202n.xlsx</code>;
> * For TERNA dataset the generation data is contained in <code>TERNA-gen-202n.xlsx</code> while the load data is in <code>TERNA-load-202n.xlsx </code>  
>
> where n goes from 1 to 5 in order to represent the five-year-period.
